```python
import os
import pandas as pd
import numpy as np

from pandas.io.json import json_normalize
# set seed for reproducibility

from gmplot import gmplot
BASE_DIR = "../"

gmap = gmplot.GoogleMapPlotter(-34.9964998500295, -64.96726512908937, 5)
dataset = pd.read_json(BASE_DIR + "input/pasos.json")
```

#### Observaciones
- Un conteo agrupado por provincias nos dice la cantidad de pasos fronterizos que hay por provincia.
- Notar que Córdoba, San Luis, Tucuman tiene al menos un paso
- Neuquen aparece repetido

```python
dataset.groupby(['province']).count()
```

- Si observamos los nombres, podremos notar que se pueden identificar que algunos estan identificados como Aeropuertos.

```python
filter_airports = dataset["name"].str.contains("AERO")
dataset[filter_airports].groupby(["province"]).count()
```

- A estos entonces los podemos filtrar para evitar que aparezcar puntos raros en el mapa

```python
lats, lons = zip(*dataset['location'])
gmap.scatter(lats, lons, 'cornflowerblue',size=2000, marker=False) 
# Draw
gmap.draw("pasos_fronterizos.html")
```
